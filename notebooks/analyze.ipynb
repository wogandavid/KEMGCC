### Fuel consumption

```python
fuel_cons = dataframes['ELWAfcon_xls']
fuel_cons = fuel_cons.set_index(['c','trun','f'])
fuel_cons = fuel_cons.pivot_table(values='Value',index='c',columns='f')
fuel_cons
```

### Capacity builds

```python
ELWAbld_xls = dataframes['ELWAbld_xls']
ELWAbld_xls.columns = ['c','trun','ELp','value']
ELWAbld_xls.head()
```

```python
ELWAbld_xls = ELWAbld_xls.pivot_table(values='value',index=['c','trun'],columns='ELp')
ELWAbld_xls
```

### Electricity capacity

```python
ELWAcap = dataframes['ELWAcap']
ELWAcap.columns = ['trun','c','ELp','sector','value']
ELWAcap = ELWAcap.drop(columns='sector')
```

```python
ELWAcap = ELWAcap.pivot_table(values='value',index=['c','trun'],columns='ELp')
ELWAcap
```

### Electricity production
